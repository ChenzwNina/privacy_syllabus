```python
#If the Google file is a docx, use 'python-docx' to extract content
from docx import Document

def extract_text_from_docx(file_path):
    # Load the .docx file
    doc = Document(file_path)
    
    # Extract text from each paragraph
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    
    # Join all paragraphs together into a single string
    return '\n'.join(full_text)
```

```python
#If the Google file is a pdf, use 'python-docx' to extract content.

import PyPDF2
from PyPDF2 import PdfFileReader

def extract_text_from_pdf(file_path):
    # Open the PDF file
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        
        # Extract text from each page
        text = ""
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    
    return text
```

```python
#If the syallbus is on a website, use bs4 to extract content.

import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def extract_text_from_webpage(url):
    links = []
    
    # Send a GET request to the webpage
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract all the text from the webpage
        content = soup.get_text()
    
        #Extract hyperlinks from syllabus and save
        for a_tag in soup.find_all('a', href=True):
            hyperlink_text = a_tag.get_text()
            href = a_tag['href']
            full_url = urljoin(url, href)  # Convert to absolute URL
            links.append((hyperlink_text, full_url))
        
        return content, links
    else:
        content = "error"
        links = ["error"]        
        return contents, links
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin

def is_external_link(href, url):
    # Check if the link starts with the base URL (indicating it's internal)
    return not href.startswith(url)

def extract_text_and_links(url):
    # Fetch the website content
    response = requests.get(url)
    
    if response.status_code == 200:
        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Iterate through all text elements in the document
        for element in soup.find_all(text=True):
            parent = element.parent
            if parent.name == "a" and 'href' in parent.attrs:
                href = parent['href']
                
                # If the link is relative, convert it to an absolute URL
                href = urljoin(url, href)
                
                # If the link is external, append the URL; otherwise, just keep the text
                if is_external_link(href, url):
                    print(f"{element} ({href})")
                else:
                    print(element)
            else:
                print(element.strip())
    else:
        print(f"Failed to retrieve content. Status code: {response.status_code}")

# Example usage
url = "https://pages.cs.wisc.edu/~cs703-1/"  
extract_text_and_links(url)
```
