````python
import requests
from bs4 import BeautifulSoup

def extract_text(url):
    # Fetch the website content
    response = requests.get(url)
    
    if response.status_code == 200:
        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extract all text from the parsed HTML, removing any HTML tags
        plain_text = soup.get_text(separator="\n")
        
        # Clean up the text, removing unnecessary blank lines and extra spaces
        cleaned_text = "\n".join([line.strip() for line in plain_text.splitlines() if line.strip()])
        
        return cleaned_text
    else:
        return f"Failed to retrieve content. Status code: {response.status_code}"

# Example usage
url = "https://pages.cs.wisc.edu/~cs703-1/"  # Replace with the actual URL you want to parse
syllabus_content = extract_text(url)

# Now, `syllabus_content` contains all the text from the webpage
print(syllabus_content)
````
